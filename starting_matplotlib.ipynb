```python
import matplotlib.pyplot as plt
%matplotlib inline
days = [1, 2, 3, 4, 5, 6, 7]
max_temp = [50, 51, 52, 48, 47, 49, 46]
min_temp = [43, 42, 40, 44, 33, 35, 37]
average_temp = [45, 48, 48, 46, 40, 42, 41]
plt.xlabel("Days")
plt.ylabel("Tempreature")
plt.title("weather")
plt.plot(days , max_temp, label="max")
plt.plot(days , min_temp, label="min")
plt.plot(average_temp, label="average")
plt.legend(loc="best")
plt.grid()
```

#BAR CHART

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Data 
company = ["sourcecode", "google", "microsoft", "insta", "FB"]
revenue = [55, 45, 33, 23, 45]

# Creating a bar chart
plt.figure(figsize=(10, 6))
plt.bar(company, revenue, color=['black', 'red', 'green', 'yellow', 'blue'])

# Adding a title and labels
plt.xlabel("Companies")
plt.ylabel("Revenue (in billions)")
plt.title("Company Annual Revenue")

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data: ages of patients
patient_ages = [23, 25, 31, 35, 45, 52, 58, 60, 63, 70, 22, 30, 34, 40, 50, 54, 57, 61, 65, 72, 
                21, 29, 33, 38, 44, 51, 55, 59, 64, 68, 27, 48, 36, 62, 41, 66, 37, 49, 43, 53]

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(patient_ages, bins=10, edgecolor='black', color='skyblue')

# Add titles and labels
plt.title('Distribution of Patient Ages', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Patients', fontsize=12)

# Adding grid for better readability
plt.grid

# Show the histogram
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data: ages of 1000 patients between 20 and 80
np.random.seed(42)
patient_ages = np.random.randint(20, 80, 1000)

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(patient_ages, bins=15, edgecolor='black', color='skyblue')

# Add titles and labels
plt.title('Distribution of Patient Ages', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Patients', fontsize=12)

# Adding grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Show the histogram
plt.show()
```
